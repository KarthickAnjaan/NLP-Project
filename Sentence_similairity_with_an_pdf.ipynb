```python
import pandas as pd
import cv2
!pip install PyPDF2
import PyPDF2
from PyPDF2 import PdfReader

# Sentence Tokenization

import nltk
nltk.download('punkt')

# SENTENCE TRANSFORMERS

!pip install -q sentence_transformers

from sentence_transformers import SentenceTransformer,util

import torch

#downloading pre-trained sentence embedded models

symmetric_embedder=SentenceTransformer('all-mpnet-base-v2')

```

```python
a= PdfReader('/content/Employee Separation Policy.pdf')
```

```python
text_1=''
for i in range(0,3):
  page = a.pages[i]
  text_1+= page.extract_text()
```

```python
text_1
```

```python
# Use NLTK's sent_tokenize method to split the text into sentences
sentences = nltk.sent_tokenize(text_1)

# Store each sentence in a variable
nlp = []
for sentence in sentences:
    nlp.append(sentence)

print(nlp)
```

```python
# Encoding the corpus

corpus_1=symmetric_embedder.encode(nlp)
```

```python
text2='The employee should not take leaves  under any situation while on the notice period'
```

```python
# Encoding the query

query_embedding=symmetric_embedder.encode(text2,convert_to_tensor=True)
```

```python
hits = util.semantic_search(query_embedding, corpus_1)
hits = hits[0]

for hit in hits:
    print(nlp[hit['corpus_id']], 'score:',round(hit['score'],4))
```

```python

# Create a list of dictionaries from the search results
hits1 = [{'corpus_id': nlp[hit['corpus_id']], 'score': round(hit['score'], 4)} for hit in hits]

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(hits1)

# Print the DataFrame
print('\n\nQuery ==>',text2)
print('-----------------------')
print('\nThe Probability of Similar sentences in the corpus:')
print('\n',df)
```
